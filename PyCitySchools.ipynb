```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
index = [0]
columns = ["Total number of Schools", "Total number of Students", "Total Budget","Average Math Scores",
     "Average Reading Scores",
     "% Passing Math",
     "% Passing Reading"]


district_summary_df = pd.DataFrame(index=index, columns=columns)


district_summary_df["Total number of Schools"] = len(school_data.index)

district_summary_df["Total number of Students"] = len(student_data.index)

district_summary_df["Total Budget"] = school_data["budget"].sum()
                                          
district_summary_df["Average Math Scores"] = student_data["math_score"].mean()

district_summary_df["Average Reading Scores"] = student_data["reading_score"].mean()

filtered_student_data_math = student_data[student_data["math_score"] >= 70]
district_summary_df["% Passing Math"] = ((filtered_student_data_math["Student ID"].count()/district_summary_df["Total number of Students"]) * 100)

filtered_student_data_math = student_data[student_data["reading_score"] >= 70]
district_summary_df["% Passing Reading"] = ((filtered_student_data_math["Student ID"].count()/district_summary_df["Total number of Students"]) * 100 )

district_summary_df["Passing Rate"] = (district_summary_df["Average Math Scores"] + district_summary_df["Average Reading Scores"])/2



district_summary_df

district_summary_df.style.format({"Total Budget": "${:,.2f}", 
                       "Average Reading Score": "{:.1f}", 
                       "Average Math Score": "{:.1f}",  
                       "Overall Passing Rate": "{:.1%}"})
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
school_summary = school_data_complete.groupby(['school_name', 'type'])

total_students = school_summary["Student ID"].count()

total_budget = school_summary["budget"].first()
per_student_budget = (total_budget/total_students)
avg_math = school_summary["math_score"].mean()

avg_read = school_summary["reading_score"].mean()

passing_math = school_data_complete.loc[school_data_complete["math_score"] >= 70]
percent_math = ((passing_math.groupby("school_name")["math_score"].count()/total_students) * 100)

passing_reading = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
percent_reading = ((passing_reading.groupby("school_name")["reading_score"].count()/total_students) * 100)
                
overall_passing_rate = (percent_math + percent_reading)/2

school_summary_df = pd.DataFrame({
"Total Students": total_students,
"Total School Budget": total_budget,
"Per Student Budget": per_student_budget,
"Average Math Score": avg_math,
"Average Reading Score": avg_read,
"% Passing Math": percent_math,
"% Passing Reading": percent_reading,
"Overall Passing Rate": overall_passing_rate
         })


top_performing_df = school_summary_df.sort_values("Overall Passing Rate", ascending=False)
top_performing_df.head()

top_performing_df.style.format({'Total Students': '{:,}', 
                          "Total School Budget": "${:,.2f}", 
                          "Per Student Budget": "${:.2f}"})
```
